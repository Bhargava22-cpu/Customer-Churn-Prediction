```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('demographic_data.csv')
print(df.head())

#Histogram for
plt.figure(figsize=(6, 4))
sns.histplot(data=df, x='PrimaryDevice')
plt.title('Distribution of PrimaryDevice')
plt.xlabel('PrimaryDevice')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(6, 4))
sns.histplot(data=df, x='Location',color='yellow')
plt.title('Distribution of Location')
plt.xlabel('Location')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(6, 4))
household_size_counts = df['HouseholdSize'].value_counts()
plt.pie(household_size_counts, labels=household_size_counts.index, autopct=lambda pct: f'{int(pct/100.*household_size_counts.sum())}')
plt.title('Distribution of HouseholdSize')
plt.show()

```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='HouseholdSize', hue='PrimaryDevice')
plt.title('Household Size Distribution by Primary Device')
plt.xlabel('Household Size')
plt.ylabel('Count')
plt.show()
```

```python
df_final = pd.get_dummies(df, columns=['Location', 'PrimaryDevice'])

dummy_cols = df_final.columns.difference(df.columns)
df_final[dummy_cols] = df_final[dummy_cols].astype(int)

df_final.head()
```

```python
df_final.to_csv('df_final.csv', index=False)
```
